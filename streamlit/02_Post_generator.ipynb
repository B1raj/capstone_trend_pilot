```python
import os
from typing import List, Dict
import openai
```

**STEP 2: Import Libraries & Configure API**

```python
from openai import OpenAI
import os

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

**STEP 3: Define Input â€“ Trending Topics**

```python
trending_topics = [
    "AI regulation in fintech",
    "Remote work productivity myths",
    "Data privacy challenges in healthcare"
]
```

**STEP 4: Define the MASTER SYSTEM PROMPT (Run Once)**

```python
MASTER_SYSTEM_PROMPT = """
You are an expert LinkedIn content strategist and copywriter.

Your goal is to generate authentic, high-engagement LinkedIn posts
that maximize reach and interaction while avoiding promotional suppression.

Optimize for:
- Authentic storytelling over promotion
- Strong first-sentence hooks
- Specific details (numbers, moments, experiences)
- Personal insights or transformations

Hard constraints:
- Target length: 100â€“200 words
- NO external links
- Minimal promotional language
- Tone: human, reflective, conversational
- Platform: LinkedIn (B2B audience)

Output should feel like a real professional sharing a genuine insight.
"""
```

**STEP 5: Create the USER PROMPT BUILDER (Key Step)**

```python
def build_user_prompt(
    topic: str,
    personal_context: str,
    insight: str,
    hook_style: str
) -> str:

    return f"""
Create a high-performing LinkedIn post using the following inputs.

Topic:
{topic}

Core personal insight or learning:
{insight}

Personal context:
{personal_context}

Hook preference:
{hook_style}

Constraints:
- 100â€“200 words
- No external links
- No promotional language
- Short paragraphs
- End with a thoughtful question
"""
```

**STEP 6: Call the LLM to Generate ONE Post**

```python
def generate_post(system_prompt: str, user_prompt: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # fast + cheap + good for generation
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content

```

**STEP 7: Test With a Single Topic (IMPORTANT CHECK)**

```python
test_prompt = build_user_prompt(
    topic="AI regulation in fintech",
    personal_context="Working as a data professional in financial services",
    insight="I realized regulation isnâ€™t slowing innovationâ€”itâ€™s reshaping it",
    hook_style="I used to believeâ€¦ now I realizeâ€¦"
)

post = generate_post(MASTER_SYSTEM_PROMPT, test_prompt)
print(post)
```

```python
HOOK_STYLES = [
    "Contrarian",
    "Personal transformation",
    "Hidden insight nobody talks about"
]

def generate_variants_for_topic(topic: str) -> List[Dict]:
    variants = []

    for hook in HOOK_STYLES:
        prompt = build_user_prompt(
            topic=topic,
            personal_context="Reflecting on my recent work and conversations",
            insight="My perspective on this topic changed after seeing it play out in practice",
            hook_style=hook
        )

        post_text = generate_post(MASTER_SYSTEM_PROMPT, prompt)

        variants.append({
            "topic": topic,
            "hook_style": hook,
            "post_text": post_text,
            "word_count": len(post_text.split())
        })

    return variants
```

```python
all_generated_posts = []

for topic in trending_topics:
    topic_variants = generate_variants_for_topic(topic)
    all_generated_posts.extend(topic_variants)
```

```python
for post in all_generated_posts:
    print("=" * 80)
    print(f"Topic: {post['topic']}")
    print(f"Hook: {post['hook_style']}")
    print(f"Word Count: {post['word_count']}")
    print(post["post_text"])
```
